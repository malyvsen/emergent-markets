```python
import numpy as np
import plotly.graph_objects as go
```

```python
with open("./hyperparameter_results.json") as results_file:
    results = [eval(line[:-2]) for line in results_file.readlines()[1:]]
len(results)
```

```python
mean_balance = lambda result: np.mean(result["total_balances"])
worst_balance = lambda result: min(result["total_balances"])
max(worst_balance(result) for result in results)
```

```python
go.Figure(
    layout=dict(
        xaxis=dict(title="Learning rate", type="log"),
        yaxis=dict(title="Final total balance", type="log"),
    ),
    data=[
        go.Scatter(
            name="Mean",
            mode="markers",
            x=[result["hyperparameters"]["learning_rate"] for result in results],
            y=[mean_balance(result) for result in results],
        ),
        go.Scatter(
            name="Minimum",
            mode="markers",
            x=[result["hyperparameters"]["learning_rate"] for result in results],
            y=[worst_balance(result) for result in results],
        ),
    ]
)
```
